```python
#how='ineer' = inner/natural join
#how='outer' = full outer join
#how='left' = left outer join
#how='right' = right outer join

import pandas as pd
```

```python
#1st table
df1 = pd.read_csv(r"C:\Users\jassm\OneDrive\Desktop\Data_Analyst_Bootcamp\Pandas\LOTR.csv")
df1
```

```python
#2nd table
df2 = pd.read_csv(r"C:\Users\jassm\OneDrive\Desktop\Data_Analyst_Bootcamp\Pandas\LOTR 2.csv")
df2
```

```python
#left/1st table = 'df1', right/2nd tabe = 'df2'
df1.merge(df2)
#output is inner join by default
```

```python
##df1.merge(df2, how='inner', on = ['FellowshipID')
#when joining 'on' one specific column, it seperates the tables
df1.merge(df2, how='inner', on = ['FellowshipID', 'FirstName'])
#joining 'on' two specific columns
```

```python
df1.merge(df2, how='outer')
#overlapping data not duplicated
```

```python
df1.merge(df2, how='left')
```

```python
df1.merge(df2, how='right')
```

```python
df1.merge(df2, how='cross')
#takes each value from left and compares to right
```

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_left', rsuffix = '_right')
```

```python
df3 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_left', rsuffix = '_right')
df3
#'Fellowship' is the index
```

```python
#merge function more smoother than the join function
```

```python
#concatenate function putting one df ontop of on another instead of on the side like merg/join function
pd.concat([df1,df2], join= 'inner')
```

```python
pd.concat([df1,df2], join= 'outer', axis= 1)
```
